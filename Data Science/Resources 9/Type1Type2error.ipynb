# Hypothesis - 

Hypotheses tests are basically tests where you take a claim and put it on trail by assessing the evidence against it.
If these is sufficient evidence against it, you reject it, but if there is insufficient evidence against it, you accept it.

You may correctly accept or reject the null hypotheses, but even considering the evidence, it is also possible to make an error. You may reject a valid null hypotheses, or you might accept it when it is actually false.

Where Type 1 and Type 2 errors become false convictions and mistaken acquittals.

Statisticians have special names for these types of errors.

1. Type 1 error - called alpha - when you wrongly reject a true null hypothesis.
2. Type 2 error - called beeta - when you worngly accept a false null hypothesis.
