## üîÑ Nested Loops

A **nested loop** is a loop inside another loop. You can nest `for` loops, `while` loops, or a combination of both.

Useful for:
- Tables and matrices
- Multi-level data (like students and subjects)
- Grids or patterns


### üî¢ Multiplication Table (3x3)

```python
# Print 3x3 multiplication table
for i in range(1, 4):
    for j in range(1, 4):
        print(i * j, end="\t")
    print()
```

### ‚≠ê Right-Angled Triangle Pattern

```python
# Print a right-angled triangle of *
rows = 5
for i in range(1, rows + 1):
    for j in range(i):
        print("*", end=" ")
    print()
```

### üßÆ Matrix Sum without List

```python
# Sum elements in a 3x3 matrix using nested loops and user input
total = 0
for i in range(1, 4):
    for j in range(1, 4):
        num = int(input(f"Enter element at position [{i},{j}]: "))
        total += num

print("Total sum of all matrix elements is:", total)
```

## üõë `break` Statement

The `break` statement is used to **exit the loop immediately**, even if the condition is still `True`.

Useful when:
- You find what you‚Äôre looking for
- You want to stop early


```python
for i in range(1, 101):
    if i % 7 == 0:
        print("First multiple of 7 is:", i)
        break
```

```python
# Example: Stop when user enters a negative number
while True:
    num = int(input("Enter a number (negative to stop): "))
    if num < 0:
        print("Negative number entered. Exiting loop.")
        break
    print("You entered:", num)
```

## ‚è≠Ô∏è `continue` Statement

The `continue` statement **skips the current iteration** and moves to the next.

Useful when:
- You want to ignore certain cases
- Filtering or skipping values


```python
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

```python
for i in range(1, 11):
    if i == 5:
        break
    print(i)
```

### Another example of continue

```python
# Example: Skip even numbers
for i in range(1, 10):
    if i % 2 == 0:
        continue
    print("Odd number:", i)
```

## üéì Case Study: Student Attendance Monitoring (Education Domain)

**Problem Statement:**
A class has 5 students. For each student, ask how many days they were present out of 30.

- Skip the entry if the attendance is negative (use `continue`)
- If attendance is more than 30, show a warning and stop the loop (use `break`)
- Use a nested loop structure if we later expand to include multiple classes


```python
for student in range(1, 6):
    print(f"\nChecking Student {student}")
    attendance = int(input("Enter attendance (0 to 30): "))

    if attendance < 0:
        print("Invalid data. Skipping student.")
        continue

    if attendance > 30:
        print("Data error. Stopping system.")
        break

    print("Recorded attendance:", attendance)
```
