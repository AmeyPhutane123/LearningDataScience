1. Write a Python program to simulate the following scenarios:

a. Tossing a coin 10,000 times and calculating the experimental probability of heads and tails.

```python
import random

# Number of tosses
tosses = 10000
heads = 0
tails = 0

for _ in range(tosses):
    result = random.choice(["H", "T"])
    if result == "H":
        heads += 1
    else:
        tails += 1

# Experimental probabilities
prob_heads = heads / tosses
prob_tails = tails / tosses

print("Coin Toss Simulation")
print("Heads:", heads, "Probability:", prob_heads)
print("Tails:", tails, "Probability:", prob_tails)
```

b. Rolling two dice and computing the probability of getting a sum of 7.

Steps:

a. Use Python's random module for simulations.

b. Implement loops for repeated trials.

c. Track outcomes and compute probabilities.

```python
import random

trials = 10000
sum_7_count = 0

for _ in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)

    if die1 + die2 == 7:
        sum_7_count += 1

prob_sum_7 = sum_7_count / trials

print("\nTwo Dice Simulation")
print("Number of times sum = 7:", sum_7_count)
print("Probability of getting sum 7:", prob_sum_7)
```

2. Write a function to estimate the probability of getting at least one "6" in 10 rolls of a fair die.

Steps

a. Simulate rolling a die 10 times using a loop.

b. Track trials where at least one "6" occurs.

c. Calculate the proportion of successful trials.

```python
import random

def probability_at_least_one_six(trials=10000):
    success = 0

    for _ in range(trials):
        found_six = False

        for _ in range(10):  # Roll die 10 times
            if random.randint(1, 6) == 6:
                found_six = True
                break

        if found_six:
            success += 1

    return success / trials


# Call the function
probability = probability_at_least_one_six()
print("\nProbability of getting at least one 6 in 10 rolls:", probability)
```

# Conditional Probability and Bayes' Theorem


3. A bag contains 5 red, 7 green, and 8 blue balls. A ball is drawn randomly, its color noted, and it is put back into the bag. If this process is repeated 1000 times, write a Python program to estimate:  
  a. The probability of drawing a red ball given that the previous ball was blue.  
  b. Verify Bayes' theorem with the simulation results.  


Steps  
    a. Use random sampling to simulate the process.  
    b. Compute conditional probabilities directly from the data.


```python
import random

# Bag composition
bag = ["R"] * 5 + ["G"] * 7 + ["B"] * 8

trials = 1000
previous = None

count_prev_blue = 0
count_red_given_blue = 0

# Simulation
for _ in range(trials):
    current = random.choice(bag)

    if previous == "B":
        count_prev_blue += 1
        if current == "R":
            count_red_given_blue += 1

    previous = current
```

a. The probability of drawing a red ball given that the previous ball was blue.

```python
if count_prev_blue > 0:
    prob_red_given_blue = count_red_given_blue / count_prev_blue
else:
    prob_red_given_blue = 0

print("P(Red | Previous Blue) =", prob_red_given_blue)
```

b. Verify Bayes' theorem with the simulation results.

```python
count_red = 0
count_blue = 0
count_blue_given_red = 0

previous = None

for _ in range(trials):
    current = random.choice(bag)

    if current == "R":
        count_red += 1
    if current == "B":
        count_blue += 1

    if previous == "R" and current == "B":
        count_blue_given_red += 1

    previous = current

# Probabilities
P_R = count_red / trials
P_B = count_blue / trials

if count_red > 0:
    P_B_given_R = count_blue_given_red / count_red
else:
    P_B_given_R = 0

# Bayes calculation
bayes_result = (P_B_given_R * P_R) / P_B

print("Bayes Theorem Result =", bayes_result)
```

# Random Variables and Discrete Probability


4. Generate a sample of size 1000 from a discrete random variable with the following distribution:  
  - P(X=1) = 0.25  
  - P(X=2) = 0.35  
  - P(X=3) = 0.4  
  Compute the empirical mean, variance, and standard deviation of the sample.  
  Steps  
      a. Use numpy.random.choice() to generate the sample.  
      b. Use numpy methods to calculate mean, variance, and standard deviation.


Step 1: Import NumPy

```python
import numpy as np
```

Step 2: Generate the discrete random sample

```python
# Possible values of X
values = [1, 2, 3]

# Given probabilities
probabilities = [0.25, 0.35, 0.4]

# Sample size
sample_size = 1000

# Generate sample
sample = np.random.choice(values, size=sample_size, p=probabilities)
```

Step 3: Compute empirical statistics

```python
# Empirical mean
mean = np.mean(sample)

# Empirical variance
variance = np.var(sample)

# Empirical standard deviation
std_dev = np.std(sample)

print("Empirical Mean:", mean)
print("Empirical Variance:", variance)
print("Empirical Standard Deviation:", std_dev)
```

# Continuous Random Variables


5. Simulate 2000 random samples from an exponential distribution with a mean of 5. Visualize the distribution using:  
  a. A histogram.  
  b. A probability density function (PDF) overlay.  
  Steps  
      a. Use numpy.random.exponential().  
      b. Use matplotlib to create visualizations.


Step 1: Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Step 2: Generate samples from exponential distribution

```python
# Parameters
mean = 5
sample_size = 2000

# Generate samples
samples = np.random.exponential(scale=mean, size=sample_size)
```

Step 3: Plot Histogram and PDF Overlay

```python
# Histogram
plt.hist(samples, bins=30, density=True, alpha=0.6)

# PDF of exponential distribution
x = np.linspace(0, max(samples), 1000)
pdf = (1 / mean) * np.exp(-x / mean)

# Overlay PDF
plt.plot(x, pdf)

# Labels and title
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Exponential Distribution (Mean = 5)")

plt.show()
```

# Central Limit Theorem


6. Simulate the Central Limit Theorem by following these steps  
  a. Generate 10,000 random numbers from a uniform distribution.  
  b. Draw 1000 samples of size n = 30.  
  c. Calculate and visualize the distribution of sample means.  
  Steps  
      a. Use numpy.random.uniform().  
      b. Plot both the uniform distribution and the sample mean distribution for comparison.


Step 1: Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Step 2: Generate 10,000 uniform random numbers

```python
uniform_data = np.random.uniform(low=0, high=1, size=10000)
```

Step 3: Draw 1000 samples of size 30 and compute sample means

```python
sample_size = 30
num_samples = 1000

sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(uniform_data, size=sample_size)
    sample_means.append(np.mean(sample))
```

Step 4: Plot the Uniform Distribution

```python
plt.figure()
plt.hist(uniform_data, bins=30, density=True)
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

Step 5: Plot the Distribution of Sample Means

```python
plt.figure()
plt.hist(sample_means, bins=30, density=True)
plt.title("Distribution of Sample Means (n = 30)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```
