# Machine Learning

- method of teaching computers to make decisions or predictions based on data, without explicitly programming every rule.

Key Difference
- Human writes explicit rules/logic.
- ML: Algorithm discovers patterns automatically.

### Types of Machine Learning

1. Supervised Learning - 
- uses labeled data (input output pairs) to train models, like teacher-student setup.
- Here, the model learns patterns from past data to predict future outcomes.
    1. Regression - 
        - Liner
        - Polynomial
    2. Decision Trees.
    3. Random Forests.

2. Unsupervised Learning - 
- No labels here! The model finds hidden patters in data, like a detetive solving mystery.
    1. Clustering and Dimensionality - 
        - SVD
        - PCA
        - K-means

3. Reinforcement Learning - 
- Learn by trail and error

4. Semi-Supervised Learning - 
- Mixes a few labeled examples with tons of unlabeled data - like learning with a few flashbards and big textbooks.



### Key ML Terminologies 

1. Feature - Input Variable
2. Target - What you want to predict.
3. Label - Known target in supervised learning.
4. Training Data - Data used to build the model.
5. Testing Data - Data used to evaluate the model.
6. Model - The learned formula or rules.




### Machine Learning Workflow 

ML Pipeline - 

- Data -> Cleaning -> Feature Selection -> Train/Test Split -> Modeling -> Evaluation -> Deployment

### What makes a good Model ?

- Accuracy - does the model make good predictions.
- Bias vs Variance:
    - High bias -> Underfitting (too simple).
    - High Variance -> Overfitting (too complex).
    

# 1. Linear Regression - Concept 

1. Linear:
- Means progressing from one stage to another in a single series of steps (sequential) extending along a straight or nearly straight line.
2. Regression: 
- The statistical terminology for regression is to determine the relationship between one dependent variable (can be many) and series of independent variable(only one which we are trying to predict).

#### Purpose -
- to predict or forecast value based on input.
- input is termed as independant variable and output as dependent variable.

#### Types of Linear Regression - 

The linear regression line has an equation: 
- y = Beeta0 + Beeta1*x
- y = mx + c
- where Y is dependent variable, X is independent variable, Beeta1 or m is slope and Beeta0 or c is intercept (the value of y when x = 0).

There are 2 types of Linear Regression models: 

1. Simple Linear Regression - 
- Is is a linear regression model with one independent variable and one dependent variable y = B0 +B1*x1

2. Multiple Linear Regression - 
- It is linear regression model with more than one independent variable and one dependent variable, y = B0 + B1*x1 + B2*x2 .....

#### Linear Regression Assumptions 
1. Linearity - linear relationship between Y and each x
2. Homoscedasticity - residuals have constant variance across all  predicted values.
3. Multivariate Normality - Residuals should be normally distributed around the regression line.
4. Independence - of the observations, includes "no autocorrction"
5. Lack of Multicollinearity - Predictiors are not correlated with each other.
6. The Outlier Check - Extreme values that may disproportionately influence the regression results.

-- If p value for any feature is < 0.05, that variable is important, and suggest not drop from dataset.


#### R - Squared
- tells us how well the model fits the data. Closer to 1 = better fit.
- Example: If R^^2 = 91, the model explains 91% of the variation in y.
- But R^^2 has a issue - it always increases when we add more features, even useless ones!

### Adjusted R-squared deals with this issue.
- fixes this quirk by penalizing extra features that don't help.
- Rule of thumb: If a feature doesn't improve the model enough, Adjusted R^^2 decreases.
