# Statistics for Data Science:

### What is Statistics?
- Statistics is the science of collecting, organizing, analyzing, and interpreting data to make informed decisions.

### Why Statistics?
- To extract insights from raw data.
- To summarize and understand data behavior.
- To make predictions and data-driven decisions.


# Descriptive Statistics:

1. Mean: The average value of a dataset.
    - Calculated by summing all values and dividing by the number of values.
    - When to Use: Use mean for symmetric distributions without outliers.
2. Median: The middle value when the data is sorted.
    - The middle value when all data points are arranged in order.
    - If even number of values, it's the average of the two middle values.
    - When to Use: Use median for skewed distributions or when outliers are present or is skewed.
3. Mode: The most frequently occurring value in a dataset.
    - A dataset can have more than one mode.
    - When to Use: Use mode for categorical data or to identify the most common value.
4. Standard Deviation: A measure of the amount of variation or dispersion in a dataset.
    - Calculated as the square root of the variance.
    - Formula: SD = sqrt( Σ (xi - mean)² / N )
    - xi is each value, mean is the average, N is the number of values.
    - When to Use: Use standard deviation to understand the spread of data around the mean.
    - To understand consistency of values.
5. Variance: The average of the squared differences from the Mean.
    - Calculated by taking the average of the squared differences from the mean.
    - Formula: Var = Σ (xi - mean)² / N
    - represents spread but in squared units.
    - When to Use: Use variance to measure how far a set of numbers are spread out from their average value.
    - When standard deviation is 4, variance is 16 (4^2).
6. Range: The difference between the maximum and minimum values in a dataset.
    - Calculated as: Range = Max - Min
    - When to Use: Use range to get a quick sense of the spread of the data.
7. Percentiles & Quartiles:
    - Percentiles: Indicate the value below which a given percentage of observations fall.
    - Quartiles: Divide the data into four equal parts.
        - Q1 (25th percentile): 25% of data below this value.
        - Q2 (50th percentile/Median): 50% of data below this value.
        - Q3 (75th percentile): 75% of data below this value.
    - When to Use: Use percentiles and quartiles to understand the distribution of data and identify outliers.

 #### Understanding statistincs is crucial for data analysis as it helps summarize and interpret data effectively.




```python
# Descriptive Statistics in Python – Practical Tutorial

# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Creating a sample DataFrame for demonstration
data = {
    'Income': [25000, 28000, 30000, 29000, 27000, 100000, 26000],
    'Car_Color': ['Red', 'Red', 'Blue', 'Black', 'Red', 'Blue', 'Black']
}
df = pd.DataFrame(data)
df.head()
```



### Mean


```python
mean_income = df['Income'].mean()
print("Mean Income:", mean_income)
```

### Median

```python
median_income = df['Income'].median()
print("Median Income:", median_income)

```

### Mode

```python
mode_color = df['Car_Color'].mode()[0]
print("Most Common Car Color:", mode_color)
```

### Standard Deviation & Variance

```python
std_income = df['Income'].std()
var_income = df['Income'].var()
print("Standard Deviation:", std_income)
print("Variance:", var_income)
```

### Range

```python
range_income = df['Income'].max() - df['Income'].min()
print("Range of Income:", range_income)
```

### Percentiles & Quartiles


```python
q1 = df['Income'].quantile(0.25)
q2 = df['Income'].quantile(0.50)
q3 = df['Income'].quantile(0.75)
print("Q1 (25%):", q1)
print("Q2 (50% - Median):", q2)
print("Q3 (75%):", q3)

```

### Boxplot for Visual Summary

```python
plt.figure(figsize=(8, 4))
sns.boxplot(y=df['Income'])
plt.title('Boxplot of Income')
plt.show()
```

### Histogram for Distribution

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Income'], kde=True)
plt.title('Income Distribution')
plt.show()
```
