# Inferential Statistics

Making Conclusions Beyond the Data!

Why Inferential Statistics?
- We cannot survey every individual in a population.
- We use samples to make inferences about the larger population.
- Helps in decision-making and predictions.

2 Terms to Know:
1. Population: The entire group we want to study.
2. Sample: A subset of the population used to make inferences.

#### Sampling Methods:
1. Random Sampling: Every individual has an equal chance of being selected.
    - Randomly select samples from the population.
    - Simple and unbiased.
    - Example: Using a random number generator to select individuals.

2. Stratified Sampling: Population divided into subgroups (strata) and samples taken from
    each stratum.
    - Divide the population into distinct subgroups (strata) based on certain characteristics.
    - Take a random sample from each stratum.
    - Ensures representation from all subgroups.
    - Example: Dividing a population by age groups and sampling from each group.

3. Cluster Sampling: Population divided into clusters, some clusters are randomly selected,
    and all individuals in those clusters are studied.
    - Divide the population into clusters (e.g., geographical areas).
    - Randomly select some clusters.
    - Purpose: Cost-effective and effort when population is large and spread out.
    - Example: Selecting certain cities and surveying all residents in those cities.    

Key DIfferences:
- Random Sampling: Simple random selection from the entire population.
- Stratified Sampling: 
    - Strata are homogeneous groups.(similar characteristics within each group).
- Cluster Sampling: Focuses on entire clusters for efficiency.
    - Clusters are heterogeneous groups.(diverse characteristics within each group).

#### Example Scenario:
- TO compare or understand customer statisfaction levels across different states in India.
    - Best Method: Stratified Sampling
        - Ensures representation from each state.
        - Results can be compared across states effectively.
        - Sample is more balaned and reliable.

    - Why not Random Sampling?
        - May not represent all states equally.
        - Some states might be underrepresented or overrepresented.
        - You may get customers only from urban areas, missing rural perspectives.
        - which may leed to biased results.
        
    - Why not Cluster Sampling?
        - May lead to bias if selected clusters are not representative of the entire population.
        - Less diversity in the sample.
        - Fast and cheaper, but not good for understanding variation across all states.

### Cental Limit Theorem (CLT)

    - Key Idea: The distribution of sample means approaches a normal distribution as the sample size increases, regardless of the population's distribution.
    - No matter the shape of the population distribution, the sampling distribution of the mean will be approximately normal if the sample size is large enough (usually n > 30).

    - Key Statement:
        - When you take many random samples from a population and calculate their means, the distribution of those means will be normal (bell-shaped) as the sample size increases.
        - Even if the original population is not normally distributed, the sample means will be.
        - Provided sample size is large enough.
    
    - Population size doesn't matter. Simple means become normal when sample size is big enough. Thats the magic of CLT.

What CLT really says?
    1. Applied to distribution of sample means.
    2. Holds regardless of population distribution shape.
        - Normal, skewed, uniform, etc.
    3. Big enough sample size depends on population.
        - For normal populations, even small samples (n < 30) work.
        - For heavily skewed populations, larger samples (n > 30) are needed.
        - No universal sample size fits all.
        - But n > 30 is a common rule of thumb.

When it fails?
    1. Extreme outliers or heavy tails
        - Populations with very long tails or extreme outliers may require much larger n.
        - Sample mean may not converge quickly to normality.
    2. Small samples from non-normal populations
        - If population is highly skewed or non-normal, small samples (n < 30) may not yield normal sample means.
        - You should use non-parametric methods or bootstrap methods instead.

Confidence Intervals (CI) - Concept vs Point estimation 
    - Range of values where population mean likely lies.
    - "We are 95% confident"
    - Confidence Interval = Sample Mean ± Margin of Error
    - Margin of Error depends on:
        - Standard deviation (σ)    
        - Sample size (n)
        - Confidence level (e.g., 90%, 95%, 99%)

Margin of Error (ME) & Standard Error (SE)
    - ME: Max expected difference between sample mean and population mean.
        - Critical Value * Standard Error
    - SE: Standard deviation of sample means.
        - σ / √n
        - σ: Population standard deviation (or sample std dev if unknown)   
        
    - ME = Z * SE
        - Z: Z-score for confidence level (e.g., 1.96 for 95% CI)
        - SE = σ / √n
            - σ: Population standard deviation (or sample std dev if unknown)
            - n: Sample size
    - Smaller samples = larger margin of error.

Z Distribution vs T Distribution
    - Z Distribution:
        - Used when population std dev (σ) is known.
        - Sample size is large (n > 30).
        - Example: National census data.
    - T Distribution:
        - Used when population std dev (σ) is unknown.
        - Sample size is small (n < 30).
        - Uses t-distribution instead of z-distribution.
        - Example: Ola Cab feedback from 20 customers.

Why 95% CI = 1- alpha?
    - Commonly used confidence level.

What is Alpha (α)?
    - Significance level = 1 - Confidence Level
    - For 95% CI, α = 0.05
    - Represents probability of error you are willing to accept.
    - It is the chance that your CI does not contain the true population parameter.
    - Represents probability of Type I error (rejecting true null hypothesis)    
    - Example: α = 0.05 means 5% chance of error.
    - it is basically risk of being wrong.

What is (1 - α)?
    - Confidence Level
    - Probability that CI contains true population parameter.
    - For 95% CI, (1 - α) = 0.95
    - Represents reliability of estimation method.
    - Example: 95% CI means 95% of such intervals will contain the true mean.
    - it is basically confidence of being right.

```python
# example

import numpy as np

population = np.arange(1, 10001) # Population of 10,000 individuals
sample = np.random.choice(population, size=1000, replace=False)
print(sample[:5]) #everytime we get random 5 values
```

```python
# Central Limit Theorem Demonstration
import numpy as np  
import matplotlib.pyplot as plt

means = []
for i in range(1000):
    sample = np.random.choice(population, size=50, replace=False)
    means.append(np.mean(sample))

plt.hist(means, bins=30, edgecolor='black')
plt.title('Sampling Distribution of the Mean (n=50)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
```

```python
# Confidence Interval Demonstration
import scipy.stats as stats
import numpy as np

sample_size = 100
sample = np.random.choice(population, size=sample_size, replace=False)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # Sample standard deviation
confidence_level = 0.95
degrees_freedom = sample_size - 1
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, loc=sample_mean, scale=sample_std/np.sqrt(sample_size))
print(f"Sample Mean: {sample_mean}")
print(f"95% Confidence Interval: {confidence_interval}")
```

```python
# Example of Bell Shape Curve explaining all concepts together
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
population = np.random.normal(loc=50, scale=10, size=10000)  # Normal distribution
sample_size = 100
sample = np.random.choice(population, size=sample_size, replace=False)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
confidence_level = 0.95
degrees_freedom = sample_size - 1
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, loc=sample_mean
, scale=sample_std/np.sqrt(sample_size))
# Plotting
plt.hist(population, bins=50, density=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, np.mean(population), np.std(population))
plt.plot(x, p, 'k', linewidth=2)
plt.axvline(sample_mean, color='r', linestyle='dashed', linewidth=2)
plt.axvline(confidence_interval[0], color='b', linestyle='dashed',
    linewidth=2)
plt.axvline(confidence_interval[1], color='b', linestyle='dashed', linewidth=2)
plt.title('Population Distribution with Sample Mean and Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
print(f"Sample Mean: {sample_mean}")

```
